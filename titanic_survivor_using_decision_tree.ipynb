```python
import pandas as pd
```

```python
df = pd.read_csv(r'titanic.csv')
```

```python
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
y_net = df['Survived']
```

```python
y_net.head()
```

```python
df.drop('Survived',axis='columns',inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
df['Age_n']=encoder.fit_transform(df['Age'])
df['Sex_n']=encoder.fit_transform(df['Sex'])
df.head()
```

```python
input = df.drop(['Age','Sex'],axis='columns')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test,x_train,y_test,y_train = train_test_split(input,y_net,test_size=0.2)
```

```python
x_test.head()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier(criterion='gini')
model = model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[3,12.0000,56,0]])
```
